### Pareto Teams - Knapsack Cost - Restaurants

```python
from paretoKnapsackRestaurants import *
from utils import *
import matplotlib.cm as cm

#Import Yelp dataset
data_path = 'datasets/pickled_data/yelp/yelp_phoenix_'
    
#Import pickled data
with open(data_path + 'ids.pkl', "rb") as fp:
    phoenix_ids = pickle.load(fp)

with open(data_path + 'sim.pkl', "rb") as fp:
    phoenix_simMatrix = pickle.load(fp)

#Cost values to assign to experts
sampleCostVals = [i for i in range(5, 101, 5)]
phoenix_item_costs = [sampleCostVals[np.random.randint(low=0, high=len(sampleCostVals))] for i in range(len(phoenix_ids))]

numItems = 100 #len(phoenix_ids)
```

```python
paretoPhoenix = paretoKnapsackRestaurants(n_items=phoenix_ids[:numItems],
                                          costs=phoenix_item_costs[:numItems],
                                          simMatrix=phoenix_simMatrix[:numItems][:numItems],
                                          budget=100)
```

```python
prunedBudgets, prunedobjectives, cost_objective_map, runTime = paretoPhoenix.prefixParetoGreedy_2Guess()
```
