### Influence Maximization Datasets Preprocessing

```python
import sys, os, pickle
import pandas as pd
import numpy as np
from collections import defaultdict
import math, itertools
import networkx as nx
```

```python
data_path_PHY = 'raw_data/influence/NetPHY/phy.txt'
data_path_HEPT = 'raw_data/influence/NetHEPT/hep.txt'

#Create adjacency list of Graph stored in a dictionary
def createGraph(data_path_file):
    edges = []
    with open(data_path_file) as f:
        next(f)
        for line in f:
            edge = [int(x) for x in line.rstrip('\n').split(" ")]
            edges.append(edge)

    G = nx.MultiDiGraph()
    for edge in edges:
        u, v = edge[0], edge[1]
        G.add_edge(u,v)

    neighbors = defaultdict(list)
    for u, v in G.edges():
        neighbors[u].append(v)
        neighbors[v].append(u)
    
    return G, neighbors
```

```python
G_phy, neighbors_phy = createGraph(data_path_PHY)
```

```python
def set_edge_probabilities_IC(self, p=1/100):
    for u,v in self.G.edges():
        i = 0
        while True:
            if not self.G[u][v][i]:
                break
            i += 1
        self.G[u][v][i]['weight'] = p
```
